```python
from collections import deque
with open("text/계곡탈출in1.txt", "r") as f:
    n = int(f.readline())
    matrix = [ list(map(int, f.readline().split())) for _ in range(n)]

dx = [0, 1]
dy = [1, 0]
chk_tbl = [ [0]*n for _ in range(n)]
chk_tbl[0][0] = matrix[0][0]
Q = deque([[0, 0]])
answer = []
while Q:
    x, y = Q.popleft()
    temp = 9999
    
    for i in range(2):
        nx = x + dx[i]
        ny = y + dy[i]
        if nx > n - 1 or ny > n - 1:
            continue       
        if nx <= n-1 and ny <= n-1:
            Q.append((nx,ny))
#             print(nx,ny, end = "   ")            
            chk_tbl[nx][ny] = chk_tbl[x][y] + matrix[nx][ny]
            if nx == n-1 and ny== n-1:
                answer.append(chk_tbl[nx][ny])
                

# print()
# tot = 0
# for i in chk_tbl:
#     tot += sum(i)
#     print(i)
# print(n)
# for i in matrix:
#     print(i)
# print(tot)


Q = deque([[0, 0]])
print(answer)
```
