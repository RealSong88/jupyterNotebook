```python
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np
import os
import glob # 윈도우환경의 파일시스템을 리눅스나 유닉스화함
```

```python
model = VGG16()
```

```python
# img_path="C:/Users/RealSongBook/Pictures/DCIM/100MSDCF/DSC06730.JPG"
img_path="C:/Users/RealSongBook/Pictures/kakaoImg01.jpg"
img = image.load_img(img_path, target_size=(224,224))
```

```python
img
```

```python
X = image.img_to_array(img)
```

```python
X = np.expand_dims(X, axis=0) # 3차원의 이미지를 4차원으로
```

```python
X.shape
```

```python
X = preprocess_input(X)
```

```python
features = model.predict(X)
```

```python
label = decode_predictions(features)
```

```python
label
```

```python
print("%s(%.2f%%)" %(label[0][0][1], label[0][0][2]*100))
```

```python
img_path="C:/Users/RealSongBook/Pictures/DCIM/100MSDCF"
for filename in glob.glob(os.path.join(img_path, '*.*')):
#     print(filename)
    img = image.load_img(filename, target_size=(224,224))
    X = image.img_to_array(img)
    X = np.expand_dims(X, axis=0)
    X = preprocess_input(X)
    features = model.predict(X)
    label = decode_predictions(features)
    print(filename,"=============>","%s(%.2f%%)" %(label[0][0][1], label[0][0][2]*100))
```
