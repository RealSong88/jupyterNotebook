```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
df=pd.DataFrame(np.random.randint(60,101, size=(10, 4)),\
                columns=['국어','영어', '수학', '과학'])
```

```python
df
```

```python
df['합계'] = df.국어 + df.영어 +df.수학 +df.과학
df
```

```python
df['평균'] = df.합계 / 4
df
```

```python
df['랭킹'] = df['합계'].rank(method='min', ascending=False)
df
```

```python
df.index=['김계란','말왕','키다리형','빅마이크','용현','제시카','토르','에디홀','강호동','메롱']
df
```

```python
# 첫번째 방법
# <== 오리지널 값이 변하지않음 #inplace가 True 이면 오리지널값에 반영한다.
df.drop(columns='합계', axis=0, inplace=True)
df
```

```python
# 두번째 방법
# 오리지널 df에 어싸인 시킴
df = df.drop(columns='랭킹', axis=0) 
df
```

```python
df['sum']= df.sum(axis=1)
df
```

```python
df.sum
```

```python
df.drop(columns='sum', axis=0, inplace=True)
```

```python
df['합계'] = df.sum(axis=1)
df
```

```python
df['예체능'] = [np.random.randint(60,101) for _ in range(10) ]
df
```

```python
df['컴퓨터'] = np.random.randint(60,101, size=(10))  # <== np의 random은 컴프리 쓸필요없음
df
```

```python
df = df[['국어','영어','수학','과학','예체능','컴퓨터', '합계']] # <== fancy indexing
```
