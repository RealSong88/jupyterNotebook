```python
pop_seoul.columns=['구별','인구수','서울시민','외국인','고령자']
# rename 보다 편리
# df1.rename(columns={"인구": "인구수", "인구.3":"서울시민","인구.6":"외국인",\
#                    "65세이상고령자" :"고령자"}, inplace=True)
```

```python
# pop_seoul.set_index('구별', inplace=True)
# pop_seoul.drop('합계', inplace=True)
pop_seoul.head()
```

```python
# 첫번째 소수점 처리방법
pop_seoul['한국인비율(%)']=np.round((pop_seoul['서울시민']/pop_seoul['인구수'])*100,\
                               decimals=2)
pop_seoul['외국인비율(%)']=round((pop_seoul['외국인']/pop_seoul['인구수'])*100, ndigits=3)
pop_seoul['고령자비율(%)']=(pop_seoul['고령자']/pop_seoul['인구수'])*100
```

```python
pop_seoul.head()
```

```python
high_age = pop_seoul.sort_values(by='고령자비율(%)', ascending=False).head(3)
high_age
```
